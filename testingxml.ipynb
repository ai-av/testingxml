```python
def extract_xml_schema(xml_file):
    tree = ET.parse(xml_file)
    root = tree.getroot()
    
    schema = {}

    def traverse(element, schema_dict):
        tag_name = element.tag
        if tag_name not in schema_dict:
            schema_dict[tag_name] = set()
        
        schema_dict[tag_name].update(element.attrib.keys())

        for child in element:
            traverse(child, schema_dict)

    traverse(root, schema)

    return {tag: list(attrs) for tag, attrs in schema.items()}

# Example usage
xml_file = "/Users/sarahwalton/Downloads/aus.001.xml"
schema = extract_xml_schema(xml_file)
print(schema)
```

```python
import csv
import xml.etree.ElementTree as ET

def get_elements_by_attribute_value(xml_file, tag_name, filter_attr, filter_value, extra_attr):
    tree = ET.parse(xml_file)
    root = tree.getroot()

    matching_elements = []
   
    for elem in root.iter():  # Iterate through all elements
        elem_tag = elem.tag.split("}")[-1]  # Remove namespace if present
        if elem_tag == tag_name:  # Check if tag matches
            if elem.get(filter_attr) == filter_value:  # Check if filtering attribute matches
                text_content = elem.text.strip() if elem.text else ""  # Cleaned text content
                extra_attr_value = elem.get(extra_attr, "N/A")  # Get extra attribute (or default "N/A")
                matching_elements.append((extra_attr_value, text_content))  # Store as a tuple

    return matching_elements

def save_to_csv(data, output_file):
    """Save extracted speaker-dialogue pairs to a CSV file."""
    with open(output_file, mode="w", newline="", encoding="utf-8") as file:
        writer = csv.writer(file)
        writer.writerow(["Speaker", "Dialogue"])  # Write header row
        writer.writerows(data)  # Write data rows

# Example usage
xml_file = "/Users/sarahwalton/Downloads/aus.001.xml"
tag_name = "said"  # The tag to search within
filter_attr = "aloud"  # The attribute to filter by
filter_value = "true"  # The specific attribute value to match
extra_attr = "who"  # The additional attribute to retrieve
output_csv = "/Users/sarahwalton/Downloads/spoken_dialogue.csv"  # Output file path

# Extract data
spoken_dialogue = get_elements_by_attribute_value(xml_file, tag_name, filter_attr, filter_value, extra_attr)

# Save to CSV
save_to_csv(spoken_dialogue, output_csv)

print(f"Data successfully saved to {output_csv}")
```
